```python
import requests

class VideoSearchEngine:
    def __init__(self, index_name="videos", host="localhost", port=9200):
        self.es = Elasticsearch([{'host': 'localhost', 'port': 9200, 'scheme': 'http'}])
        self.index_name = index_name
        self.embedder = SentenceTransformer("all-MiniLM-L6-v2")

    def setup_index(self):
        if self.es.indices.exists(index=self.index_name):
            print(f"Index '{self.index_name}' already exists. Deleting...")
            self.es.indices.delete(index=self.index_name)

        mappings = {
            "properties": {
                "title": {"type": "text"},
                "embedding": {"type": "dense_vector", "dims": 384},
                "url": {"type": "keyword"}
            }
        }

        self.es.indices.create(index=self.index_name, body={"mappings": mappings})

    def fetch_data_from_api(self):
        url = 'https://'
        headers = {
            'Content-Type': 'application/json',
            'Authorization': ''
        }

        try:
            response = requests.get(url, headers=headers)
            if response.status_code == 200:
                data = response.json()
                videos = data.get('result', [])
                return videos
            else:
                print(f"Failed to fetch data from API. Status code: {response.status_code}")
        except Exception as e:
            print(f"An error occurred while fetching data from API: {e}")

        return []

    def index_article(self, videos):
        for video in videos:
            title = video.get('title', '')
            url = video.get('videoUrl', '')

            if not title:
                print("Skipping video indexing due to missing title.")
                continue

            # Check if a video with the same content already exists
            query_body = {
                "query": {
                    "bool": {
                        "should": [
                            {"match": {"title": title}},
                            {"match": {"url": url}}
                        ]
                    }
                }
            }
            response = self.es.search(index=self.index_name, body=query_body)
            hits = response["hits"]["hits"]

            # If any video with the same content already exists, skip indexing
            if hits:
                print(f"Video with title '{title}' and URL '{url}' already exists. Skipping indexing.")
                continue

            # If the video does not exist, index it
            embedding = self.embedder.encode(title)
            doc = {"title": title, "embedding": embedding.tolist(), "url": url}
            self.es.index(index=self.index_name, body=doc)
            print(f"Indexed new video: {title}")

    def semantic_search(self, query, top_k=5):
        query_embedding = self.embedder.encode(query)
        search_body = {
            "query": {
                "script_score": {
                    "query": {"match_all": {}},
                    "script": {
                        "source": "cosineSimilarity(params.query_vector, 'embedding') + 1.0",
                        "params": {"query_vector": query_embedding.tolist()}
                    }
                }
            }
        }
        response = self.es.search(index=self.index_name, body=search_body, size=top_k)
        hits = response["hits"]["hits"]
        return [{"title": hit["_source"]["title"], "url": hit["_source"]["url"]} for hit in hits]
```
